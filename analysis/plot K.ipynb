```python
import numpy as np
import matplotlib.pyplot as plt

mean = [0.2903, 0.3985, 0.4351, 0.4356] 
std = [0.0446, 0.0558, 0.0560, 0.0511]

fig = plt.figure()
plt.rcParams.update({'font.size': 14})
x = np.arange(len(mean))

plt.errorbar([1, 5, 10, 20], mean, yerr=std)
plt.xticks([1, 5, 10, 20], [1, 5, 10, 20]) 
plt.ylim([0.2, 0.5])
plt.xlabel("Number of particles (K)", fontsize=16)
plt.ylabel("Speedup", fontsize=16)
plt.tight_layout()
plt.savefig("different_K.pdf")
```
